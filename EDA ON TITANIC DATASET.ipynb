```python
#importing label_encoder
from sklearn import preprocessing

#label_encoder object has a capability of understanding word labels
label_encoder=preprocessing.LabelEncoder()

#encode lables in column Sex
data['Sex']= label_encoder.fit_transform(data['Sex'])

data['Sex'].value_counts() #converting sex(male,female) into zeros and ones for better understanding
data
```

```python
data=data.drop(['Ticket','Fare','Cabin'],axis=1)  #droping columns which have no importance in my analysis
data
```

```python
#geeting median of the Age column

data['Age'].median()
```

```python
#akfter getting the median of Age and that value is assigning, where rows are empty in the data

data['Age']=data['Age'].fillna(value=28)
data
```

```python
#assigning values where rows are NUll in Embarked column

data['Embarked']=data['Embarked'].fillna(value='S')
data
```

```python
data.isna().sum()     #checking nulls and sum of nulls 
```

```python
#converting Embarked values into zeros and ones

label_encoder=preprocessing.LabelEncoder()

#encode lables in column Embarked
data['Embarked']= label_encoder.fit_transform(data['Embarked'])
data['Embarked'].value_counts()
data
```
